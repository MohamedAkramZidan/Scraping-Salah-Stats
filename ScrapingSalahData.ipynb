```python
import pandas as pd
```

## Scraping Each Season in CSV

```python
SeasonPlay = [17,18,19,20,21,22,23,24]
PlayerName = 'Mohamed-Salah'
id = 'e342ad68'
for i in SeasonPlay:
  url = f'https://fbref.com/en/players/{id}/matchlogs/20{i}-20{i+1}/c9/{PlayerName}-Match-Logs'
  table = pd.read_html(url,attrs={'id':f'matchlogs_20{i}-20{i+1}_9'})
  to_csv = table[0].to_csv(f'M:/DF/{i}{PlayerName}.csv')
  df = pd.read_csv(f'M:/DF/{i}{PlayerName}.csv')
  df.columns = df.iloc[0]
  df.drop(df.index[0],inplace=True)
  df = df.loc[:, ~df.columns.isna()]
  df['Season'] = f'{i}_{i+1}'
  df.index.name = None
  df.dropna(inplace=True)
  df.drop('Match Report',axis=1,inplace=True)
  path = f'E:/{PlayerName}_MathStat{i}_{i+1}.csv'
  to_csv = df.to_csv(path,index=False)
  print(f'done{i}_{i+1}')
  print(df.shape)
```

## Concat in one CSV File

```python
import glob
path = "E:/"
all_files = glob.glob(path + "/Salah*.csv")

df_list = [pd.read_csv(file) for file in all_files]
combined_df = pd.concat(df_list, ignore_index=True)

combined_df.to_csv("D:/Salah-Stat-PrimerLeague.csv", index=False)
print("Combined CSV saved successfully!")
```
